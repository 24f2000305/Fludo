```
import numpy as np
import trimesh
from trimesh.creation import cylinder, cone, box
try:
    from google.colab import files
except Exception:
    files = None

body = cylinder(radius=0.6, height=6.0, sections=64)
nose = cone(radius=0.58, height=1.8, sections=64)
nose.apply_translation([0, 0, 6.9])
fins = []
fin_proto = box(extents=[0.2, 1.0, 0.8])
T = np.eye(4); T[:3, 3] = [0.7, 0.0, 0.4]
fin_proto.apply_transform(T)
for ang in [0, np.pi/2, np.pi, 3*np.pi/2]:
    R = trimesh.transformations.rotation_matrix(ang, [0,0,1])
    f = fin_proto.copy(); f.apply_transform(R); fins.append(f)
mesh = trimesh.util.concatenate([body, nose] + fins)
mesh.export('rocket.stl')
print('Saved rocket.stl')
if files: files.download('rocket.stl')
```
